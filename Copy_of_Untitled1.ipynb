importing the Dependencies


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

```

Data collection and Framing


```python
#loading the CSV data into pandas Dataframe
heart_data=pd.read_csv("/dataset.csv")
```

0-> represents HEALTHY HEART

Splitting features and Target

```python
x=heart_data.drop(columns="target",axis=1)
y=heart_data["target"]
```

```python
print(x)
```

```python
print(y)
```

Shifting the Data into training and testing data

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Model Training

LogisticRegression

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

Model Evaluation

Accuracy_Score

```python
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print("accuracy_score of training data:",training_data_accuracy)
```

```python
x_test_prediction=model.predict(x_test)
testing_data_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print("accuracy_score of test data:",testing_data_accuracy)
```

```python
input_data=(58,0,0,100,248,0,0,122,0,1,1,0,2)
#converting data array into numpy array
inp_data_np=np.asarray(input_data)
#reshaping it to predict only for data item
inp_data_reshaped=inp_data_np.reshape(1,-1)
prediction=model.predict(inp_data_reshaped)
print(prediction)
if(prediction[0]==0):
  print("The person doesn't have a heart disease")
else:
  print("The person has heart disease")
```
